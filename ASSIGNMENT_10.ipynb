```python
ANS.1


class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_of_vehicle = average_of_vehicle

# Creating an instance of the Vehicle class
car = Vehicle("Car", 200, 30)

# Accessing the instance variables
print(car.name_of_vehicle)       # Output: Car
print(car.max_speed)             # Output: 200
print(car.average_of_vehicle)    # Output: 30




In the above code, the Vehicle class is defined with an __init__ method that takes three parameters: name_of_vehicle, max_speed, and average_of_vehicle. These parameters are used to initialize the instance variables of the class, prefixed with self. to indicate they belong to the instance.

After defining the class, an instance of the Vehicle class named car is created by passing the appropriate values to the __init__ method. Finally, we can access the instance variables of car using dot notation (car.name_of_vehicle, car.max_speed, car.average_of_vehicle) and print their values.

Feel free to modify the values passed to the Vehicle constructor to suit your requirements.



ANS.2



class Car(Vehicle):
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        super().__init__(name_of_vehicle, max_speed, average_of_vehicle)

    def seating_capacity(self, capacity):
        return f"{self.name_of_vehicle} has a seating capacity of {capacity} people."

# Creating an instance of the Car class
car = Car("Car", 200, 30)

# Calling the seating_capacity method
capacity = 5
result = car.seating_capacity(capacity)
print(result)  # Output: Car has a seating capacity of 5 people.



In the above code, the Car class is defined as a child class of the Vehicle class. The __init__ method of the Car class calls the __init__ method of the Vehicle class using the super() function to inherit and initialize the instance variables name_of_vehicle, max_speed, and average_of_vehicle from the parent class.

The seating_capacity method is defined in the Car class. It takes a capacity parameter and returns a string that includes the name of the vehicle and its seating capacity. In this case, the method returns a formatted string using f-strings.

After creating an instance of the Car class, we call the seating_capacity method on the car object, passing the desired capacity as an argument. The method returns the formatted string, which is then printed to the console.

You can customize the class and method according to your specific requirements.






ANS.3




class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        pass


class Mammal(Animal):
    def __init__(self, name):
        super().__init__(name)
    
    def speak(self):
        return "Mammal: Roar!"


class Flyable:
    def fly(self):
        pass


class Bat(Mammal, Flyable):
    def __init__(self, name):
        super().__init__(name)
    
    def speak(self):
        return "Bat: Squeak squeak!"

    def fly(self):
        return "Bat: Flying high!"


# Creating an instance of Bat
bat = Bat("Batty")

# Calling the speak method from Mammal class
print(bat.speak())   # Output: Bat: Squeak squeak!

# Calling the fly method from Flyable class
print(bat.fly())     # Output: Bat: Flying high!




In the above example, we have three classes: Animal, Mammal, and Flyable. The Animal class serves as a base class with a constructor that initializes the name attribute and a placeholder method speak().

The Mammal class is a subclass of Animal and overrides the speak() method to return "Mammal: Roar!". It inherits the name attribute from the Animal class.

The Flyable class is a separate class that defines a fly() method, which is a placeholder method.

The Bat class is the final subclass that demonstrates multiple inheritance. It inherits from both the Mammal and Flyable classes. It overrides the speak() method to return "Bat: Squeak squeak!" and the fly() method to return "Bat: Flying high!".

By creating an instance of the Bat class, we can call the overridden methods speak() and fly(). The methods are resolved according to the method resolution order (MRO) in Python, which determines the order in which classes are searched for a method.

In this example, the Bat class inherits the speak() method from the Mammal class and the fly() method from the Flyable class, demonstrating the concept of multiple inheritance.







ANS.4





class Person:
    def __init__(self, name):
        self._name = name

    def get_name(self):
        return self._name

    def set_name(self, new_name):
        self._name = new_name

# Creating an instance of the Person class
person = Person("John Doe")

# Accessing the name attribute using the getter method
name = person.get_name()
print(name)  # Output: John Doe

# Modifying the name attribute using the setter method
person.set_name("Jane Doe")

# Accessing the modified name attribute using the getter method
name = person.get_name()
print(name)  # Output: Jane Doe



ANS.5



class Animal:
    def speak(self):
        return "Animal: Roar!"


class Cat(Animal):
    def speak(self):
        return "Cat: Meow!"


# Creating instances of the Animal and Cat classes
animal = Animal()
cat = Cat()

# Calling the speak method of the Animal class
print(animal.speak())  # Output: Animal: Roar!

# Calling the speak method of the Cat class (overrides the Animal class method)
print(cat.speak())     # Output: Cat: Meow!




In the above example, we have two classes: Animal and Cat. The Animal class defines a method called speak(), which returns "Animal: Roar!".

The Cat class is a subclass of Animal and overrides the speak() method with its own implementation. In this case, the speak() method in the Cat class returns "Cat: Meow!".

By creating instances of both the Animal and Cat classes, we can call the speak() method on each object. When we call animal.speak(), the speak() method from the Animal class is invoked, returning "Animal: Roar!". However, when we call cat.speak(), the speak() method in the Cat class overrides the implementation from the Animal class, returning "Cat: Meow!".

Method overriding allows subclasses to provide their own implementation of methods inherited from their parent class, enabling customization and specialization of behavior. It promotes code reusability and facilitates polymorphism, allowing objects of different classes to respond differently to the same method call.


```
